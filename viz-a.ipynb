```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Iris data visualization

```python
data = pd.read_csv('Iris.csv')
data.head(10)
```

### relation between SepalLengthCm and SepalWidthCm of Iris (Positive relationship)

```python
sns.set_theme(style="darkgrid")

sns.lineplot(x="SepalLengthCm", y="SepalWidthCm",
             hue="Species", style="Species",
             data=data)
```

### relation between type of Iris and its average length with  average width


```python
data['avg length'] = (data['PetalLengthCm']+data['SepalLengthCm'])/2
data['avg_width'] = (data['SepalWidthCm']+data['PetalWidthCm'])/2
sns.set_theme(style="white")
g = sns.JointGrid(data=data, x="avg_width", y="avg length", space=0, ratio=17, hue="Species")
g.plot_joint(sns.scatterplot, size=data["Species"], sizes=(50, 120),
             color="c", alpha=.8, legend=True )
g.plot_marginals(sns.rugplot, height=5, color="c", alpha=.8)
```
